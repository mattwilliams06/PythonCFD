````python
rho132 = np.interp(132, xi, yi)    # find density at 132K
print(f'ρ at 132K: {rho132}')
````

# Polynomial interpolation

````python
Ti = np.array([100, 150, 200])
ρi = np.array([
    3.5562,
    2.3364,
    1.7458
])
````

````python
order = len(Ti) - 1
coeffs = np.polyfit(Ti, ρi, order)
````

````python
# coefficients a0, a1, a2
coeffs
````

````python
rho = np.polyval(coeffs, 132)
print(rho)
````

````python
rhoi = 0.
for p, a in enumerate(coeffs):
    rhoi += a*132**(2-p)
rhoi
````

````python
plt.plot(T, ρ, 'go', T, np.polyval(coeffs, T))
plt.legend(['Data', 'Interpolation'])
````

The polynomial fit succeeds for the first three points, but those were the only points that it was fitted to, and thus diverges greatly with the rest of the dataset. Let's do a polynomial fit of degree n-1 for the dataset.

````python
coeffs = np.polyfit(T, ρ, len(T)-1)
````

````python
coeffs
````

````python
rho132 = 0.
for p, a in enumerate(coeffs):
    rho132 += a*132**(len(T)-1 - p)
rho132
````

````python
np.polyval(coeffs, 132)
````

````python
np.polyval(coeffs, 990)
````

````python
plt.plot(T, ρ, 'go', T, np.polyval(coeffs, T), 'r--')
plt.legend(['Data', 'Interpolation'])
````

We can see from the temperature at 990K that the high-degree polynomial does not generalize well. Obviously the polynomial goes through every point, but what happens in between the point? We can't see this from the rendering, because each interpolated point coincides with the data points, and matplotlib is simply connecting the dots. We will create a ```linspace``` to analyze the polynomial behavior in grater detail.

````python
TT = np.linspace(100, 1000, 100)
````

````python
plt.plot(T, ρ, 'go', TT, np.polyval(coeffs, TT), 'r--')
plt.legend(['Data', 'Interpolation'])
````

We will perform a manual analysis of polynomial interpolation using matrices. The coefficient matrix contains all the known values at each exponent (0, 1, 2, 3, ..., n-1).

````python
Ti = np.array([100, 150, 200])
ρi = np.array([
    3.5562,
    2.3364,
    1.7458
])
n = len(Ti)
A = np.zeros((n,n))
A
````

````python
for i in range(n):
    A[:, i] = Ti**i
A
````

````python
coeffs = np.linalg.solve(A, ρi)
````
