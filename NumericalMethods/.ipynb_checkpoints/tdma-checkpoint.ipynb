```python
def thomas(l, u, d, b):
    n = len(b)
    x = np.zeros(n)
    for i in range(1, n):
        d[i] = d[i] - u[i-1]*l[i-1]/d[i-1]
        b[i] = b[i] - b[i-1]*l[i-1]/d[i-1]
    
    x[n-1] = b[n-1]/d[n-1]
    for i in range(n-2, -1, -1):
        x[i] = (b[i]-u[i]*x[i+1])/d[i]
    return x
```
